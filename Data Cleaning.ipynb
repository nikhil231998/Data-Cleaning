```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
# Loading the dataset
starWarsDataset = pd.read_csv('StarWars.csv', skiprows = [0, 1], header = None, names = ['RespondentID', 'watched_any', 'fan_Star_Wars', 'watched_1', 'watched_2', 'watched_3', 'watched_4', 'watched_5', 'watched_6', 'rank_1', 'rank_2', 'rank_3', 'rank_4', 'rank_5', 'rank_6', 'view_Han_Solo', 'view_Luke_Skywalker', 'view_Princess_Leia_Organa', 'view_Anakin_Skywalker', 'view_Obi_Wan_Kenobi', 'view_Emperor_Palpatine', 'view_Darth_Vader', 'view_Lando_Calrissian', 'view_Boba_Fett', 'view_C-3P0', 'view_R2_D2', 'view_Jar_Jar_Binks', 'view_Padme_Amidala', 'view_Yoda', 'Which character shot first?', 'familiar_Expanded_Universe', 'fan_Expanded_Universe', 'fan_Star_Trek_franchise', 'Gender', 'Age', 'Household Income', 'Education', 'Location (Census Region)'])
```

```python
# View the dataset and check if it equal to the source data
starWarsDataset.head(20)
```

```python
# Removing white spaces from non numeric columns white spaces
for i in range(1,9):
    starWarsDataset.iloc[:,i] = starWarsDataset.iloc[:,i].str.strip()
for i in range(15,38):
    starWarsDataset.iloc[:,i] = starWarsDataset.iloc[:,i].str.strip()
```

```python
# Doing sanity checks and typos for all columns
for i in range(0,38):
    print(starWarsDataset.iloc[:,i].value_counts())
```

```python
# Correcting typos in "fan_Star_Wars" coulmn
starWarsDataset['fan_Star_Wars'].replace('Yess', 'Yes', inplace=True)
starWarsDataset['fan_Star_Wars'].replace('Noo', 'No', inplace=True)

# Correcting typos in "fan_Expanded_Universe" coulmn
starWarsDataset['fan_Expanded_Universe'].replace('Yess', 'Yes', inplace = True)

# Correcting typos in "fan_Star_Trek_franchise" coulmn
starWarsDataset['fan_Star_Trek_franchise'].replace('no', 'No', inplace = True)
starWarsDataset['fan_Star_Trek_franchise'].replace('yes', 'Yes', inplace = True)
starWarsDataset['fan_Star_Trek_franchise'].replace('Noo', 'No', inplace = True)

# Correcting typos in "Gender" coulmn
starWarsDataset['Gender'].replace('female', 'Female', inplace = True)
starWarsDataset['Gender'].replace('male', 'Male', inplace = True)
starWarsDataset['Gender'].replace('F', 'Female', inplace = True)

# Correcting the value of age = 500 which was found during sanity checks and putting 500 in the series which has the highest mode 
starWarsDataset['Age'].replace('500', '45-60', inplace = True)
```

```python
# Checking the outcome after correcting typos and sanity checks
for i in range(0,38):
    print(starWarsDataset.iloc[:,i].value_counts())
```

```python
# Putting "Yes" if a person has watched any movie
starWarsDataset['watched_1'].replace('Star Wars: Episode I  The Phantom Menace', 'Yes', inplace=True)
starWarsDataset['watched_2'].replace('Star Wars: Episode II  Attack of the Clones', 'Yes', inplace=True)
starWarsDataset['watched_3'].replace('Star Wars: Episode III  Revenge of the Sith', 'Yes', inplace=True)
starWarsDataset['watched_4'].replace('Star Wars: Episode IV  A New Hope', 'Yes', inplace=True)
starWarsDataset['watched_5'].replace('Star Wars: Episode V The Empire Strikes Back', 'Yes', inplace=True)
starWarsDataset['watched_6'].replace('Star Wars: Episode VI Return of the Jedi', 'Yes', inplace=True)
```

```python
# Checking the new Data
starWarsDataset.head(20)
```

```python
# Deleting 100 records where a person has seen a movie but the value of rest of all the entries in blank
starWarsDataset.drop(starWarsDataset[(starWarsDataset['watched_any'] == 'Yes') & starWarsDataset['fan_Star_Wars'].isnull()].index, inplace = True)

# Deleting 10 records which have not seen any movie and do not have any value for the rest of the columns 
starWarsDataset.drop(starWarsDataset[(starWarsDataset['watched_any'] == 'No') & starWarsDataset['fan_Star_Trek_franchise'].isnull()].index, inplace = True)
```

```python
# Checking the result
starWarsDataset[(starWarsDataset['watched_any'] == 'Yes') & starWarsDataset['fan_Star_Wars'].isnull()]
starWarsDataset[(starWarsDataset['watched_any'] == 'No') & starWarsDataset['fan_Star_Trek_franchise'].isnull()]
```

```python
#Putting "No" values instead of NaN for fan_Star_Wars column
starWarsDataset.iloc[:,2].fillna("No", inplace = True)

# Putting "No" instad of NaN if a person has not watched a movie for all the movie columns
for i in range(3,9):
    starWarsDataset.iloc[:,i].fillna("No", inplace = True)
```

```python
# Changing the datatype of columns to categorical 
for i in range(1,9):
    starWarsDataset.iloc[:,i] = pd.Categorical(starWarsDataset.iloc[:,i], ['YES', 'NO'])
    
for i in range(9,15):
    starWarsDataset.iloc[:,i] = pd.Categorical(starWarsDataset.iloc[:,i], [0, 1, 2, 3, 4, 5, 6], ordered = True)
    
for i in range(15,29):
    starWarsDataset.iloc[:,i] = pd.Categorical(starWarsDataset.iloc[:,i], ['UNFAMILIAR (N/A)', 'VERY UNFAVORABLY', 
                                                                           'SOMEWHAT UNFAVORABLY', 'NEUTRAL', 
                                                                           'SOMEWHAT FAVORABLY', 'VERY FAVORABLY'],
                                               ordered = True)

starWarsDataset.iloc[:,29] = pd.Categorical(starWarsDataset.iloc[:,29], ['NA', "I DON'T UNDERSTAND THIS QUESTION",
                                                                        'GREEDO', 'HAN'])
    
for i in range(30,33):
    starWarsDataset.iloc[:,i] = pd.Categorical(starWarsDataset.iloc[:,i], ['YES', 'NO'])
    
starWarsDataset.iloc[:,33] = pd.Categorical(starWarsDataset.iloc[:,33], ['FEMALE', 'MALE'])

starWarsDataset.iloc[:,34] = pd.Categorical(starWarsDataset.iloc[:,34], ['18-29', '30-44', '45-60', '> 60'], ordered = True)

starWarsDataset.iloc[:,35] = pd.Categorical(starWarsDataset.iloc[:,35], ['NOT AVAILABLE', '$0 - $24,999', '$25,000 - $49,999',
                                                                         '$50,000 - $99,999', '$100,000 - $149,999', 
                                                                         '$150,000+'], ordered = True)

starWarsDataset.iloc[:,36] = pd.Categorical(starWarsDataset.iloc[:,36], ['LESS THAN HIGH SCHOOL DEGREE', 'HIGH SCHOOL DEGREE',
                                                                         'SOME COLLEGE OR ASSOCIATE DEGREE', 
                                                                         'BACHELOR DEGREE', 'GRADUATE DEGREE'], ordered = True)

starWarsDataset.iloc[:,37] = pd.Categorical(starWarsDataset.iloc[:,37], ['EAST NORTH CENTRAL', 'EAST SOUTH CENTRAL', 'WEST NORTH CENTRAL', 
                                                                         'WEST SOUTH CENTRAL', 'MIDDLE ATLANTIC', 'SOUTH ATLANTIC', 
                                                                         'PACIFIC', 'MOUNTAIN', 'NEW ENGLAND'])
```

```python
# Task 2: Data Exploration
# 1. Explore the survey question: \textit{Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.	(Star Wars: Episode I  The Phantom Menace; Star Wars: Episode II  Attack of the Clones; Star Wars: Episode III  Revenge of the Sith;	Star Wars: Episode IV  A New Hope; Star Wars: Episode V The Empire Strikes Back; Star Wars: Episode VI Return of the Jedi)}, then analysis how people rate Star Wars Movies. 


#Code goes after this line by adding cells
```

```python
leg_labels = ['Episode I', 'Episode II', 'Episode III', 'Episode IV', 'Episode V', 'Episode VI']
j = -2.5
for i in range(9,15):
    unique_elements, counts_elements = np.unique(np.asarray(starWarsDataset.iloc[:,i]), return_counts=True)
    plt.bar(unique_elements+0.1*j,counts_elements, width = 0.1, label = starWarsDataset.columns[i])
    j = j+1
plt.xticks = unique_elements
plt.legend(labels = leg_labels, ncol = 3)
plt.grid(color='k', linestyle='-', linewidth=1.0)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.rcParams["figure.figsize"] = (15,5)
```

```python
avg_rank = []
for i in range(9,15):
    avg_rank.append(starWarsDataset.iloc[:,i].astype(int).mean())
movie_name = ['Episode I', 'Episode II', 'Episode III', 'Episode IV', 'Episode V', 'Episode VI']
plt.bar(movie_name,avg_rank)
plt.ylabel('Rating')
plt.rcParams["figure.figsize"] = (15,5)
plt.show()
```
